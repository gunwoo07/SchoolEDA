```python
def read_data(file_name, year=2021):
    data = []
    file = ''

    with open(f'./data/{year}/{file_name}.txt', 'r') as f:
        file = f.read()

    file = file.split('\n')
    for answer in file:
        data.append(int(answer))

    return data
```

```python
data = read_data('2021-1st-mid-math', 2021)
```

```python
print(data)
```

```python
def filter_data(data):
    for i, answer in enumerate(data):
        if answer not in [1, 2, 3, 4, 5]:
            print(f'Something Founded!: {i}th answer is {answer}')
```

```python
filter_data(data)
```

```python
def count_answer(data):
    result = [0, 0, 0, 0, 0]
    
    for answer in data:
        result[answer - 1] += 1

    return result
```

```python
print(count_answer(data))
```

```python
from matplotlib import pyplot as plt
import numpy as np

def circle_graph(title, count):
    total = sum(count)
    ratio = []

    for i in range(5):
        ratio.append((count[i]/total) * 100)

    labels = ['1번', '2번', '3번', '4번', '5번']
    plt.title(title)
    plt.pie(ratio, labels=labels, autopct="%.1f%%")
```

```python
def stick_graph(title, count, color='y'):
    x = np.arange(5)
    answers = ['1번', '2번', '3번', '4번', '5번']
    values = [count[0], count[1], count[2], count[3], count[4]]

    plt.bar(x, values, color=color)
    plt.xticks(x, answers)

    plt.title(f'{title}')
    plt.xlabel('정답')
    plt.ylabel('개수')

    plt.show()
```

```python
plt.rcParams['font.family'] = 'D2Coding'
```
